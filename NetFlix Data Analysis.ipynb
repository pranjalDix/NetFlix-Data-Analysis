# Q.2. In which year the highest number of the TV shows & Movies were released ? show with Bar Graph

```python
data.dtypes
```

```python
# We can see the datatype of Release_Date is object
# Thus we need to change the data type of Release_Date
```

```python
# Function for converting to datetime format - 

from datetime import datetime

def convert_to_datetime(Release_date):
    if isinstance(Release_date, str):
        try:
            date_string = Release_date.strip()
            return datetime.strptime(date_string, "%B %d, %Y")
        except:
            return None
    else:
        return None
```

```python
data['Release'] = data['Release_Date'].apply(convert_to_datetime)
```

```python
data.head(5)
```

```python
data.dtypes  # datatype of new column Release is datetime
```

```python
data.drop(columns = ['Release_Date'], inplace=True)
```

```python
data.head()
```

```python
year_counts = data['Release'].dt.year.value_counts().sort_values(ascending=False)
year_counts
```

```python
data['Release'].dt.year.value_counts().plot(kind = 'bar')  # bar plot using pandas library
```

```python
# Horizontal bar plot using matplotlib

plt.barh(year_counts.index, year_counts.values, color="Green", edgecolor = 'Black')
plt.title("Number of Releases in each year")
plt.xlabel('Count of TV shows/Movies')
plt.ylabel('Year')
plt.show()
```

```python
print(f"{year_counts.index[0]} Year has the highest number of shows :- {year_counts.iloc[0]}")
```
